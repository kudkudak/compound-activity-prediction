```python
from data_api import *
import glob
excluded_proteins = set(["hiv_protease", "d2"])
```

```python
statistics = ["mean_mcc", "mean_wac", "mean_acc"] +  ["train_time", "test_time", "n_support"]
files = ["svm_rbf.csv","lr_K_15.csv", "lr_K_20.csv", "lr_K_30.csv", \
         "svm_linear_K_15.csv", "svm_linear_K_20.csv", "svm_linear_K_30.csv", "svm_rbf_nystr.csv", "KNN.csv", "KNN_LSH.csv", "svm_jaccard.csv"]
dfs = [[r[0:-4],pd.read_csv(r)] for r in sorted(files)]

tabelkas = {}
for s in statistics:
    D = pd.DataFrame([[v] for v in dfs[0][1]["experiment"].values], columns=["experiment"])
    for df in dfs:
        D[df[0]] = df[1][s]
    D.to_csv(s+".csv")
    svms = ["svm_linear_K_30", "svm_linear_K_20", "svm_linear_K_15"]
    lrs = ["lr_K_15", "lr_K_20", "lr_K_30"]
    D[svms] = D[svms].replace(np.NaN, 0)

    D[lrs] = D[lrs].replace(np.NaN, 0)
    D["lr"] = np.max(D[lrs].values, axis=1)
    
    D["svm_linear"] = np.max(D[svms].values, axis=1)
    D.index = D["experiment"]
    tabelkas[s] = D
```

```python
def compare(model_a=["svm_linear", "svm_linear"], model_b=["svm_lr", "svm_linear"]):
    fingerprint_statistics_3 = []
    for id,f in enumerate(fingerprints):
        fs = []
        fs.append(f)

        mean_0, mean_1 = [], []
        mean_1_cols = []
        dims2= []
        better = 0
        better_len = 0
        ## Calculate means over proteins
        for idp, p in enumerate(proteins):
            if p in excluded_proteins:
                continue

            X, Y = get_raw_training_data(fingerprint=id, protein=idp)
            dims2.append(X.shape[1])
            mean_0.append(np.mean((X > 0).todense()))
            mean_1.append(np.mean((X > 1).todense()))

            mean_1_cols.append(np.sum(np.mean((X>1).todense(), axis=0) > 0))
            results = tabelkas["mean_wac"].loc[p+":"+f]

            vals = results[model_a[0]], results[model_a[1]], results[model_b[0]], results[model_b[1]]

            if (vals[0] != np.nan or vals[1] != np.nan) and (vals[2] != np.nan or vals[3] != np.nan): #we can compare
                vals = [0 if v==np.nan else v for v in vals] # filter out nans

                better_len += 1

                if max(vals[0], vals[1]) > max(vals[2],vals[3]):
                    better += 1

        ## Append calculated statistics
        fs.append(np.array(dims2).mean())
        fs.append(np.array(mean_0).mean())
        fs.append(np.array(mean_1).mean())
        fs.append(np.array(mean_1_cols).mean())
        fs.append(better/float(better_len))
        fingerprint_statistics_3.append(fs)
    return  pd.DataFrame(np.array(fingerprint_statistics_3),\
                                          columns = ["fingeprint", "dimensions", "average>0", "average>1", \
                                                     "average no. of cols >1", "better percentage"])
```
